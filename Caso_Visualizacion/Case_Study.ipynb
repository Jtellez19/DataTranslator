```python
df = pd.read_csv("USvideos.csv")
```

```python
dfGrilla=df[:21]
imagenes=list(dfGrilla["thumbnail_link"]) #url
fig, axs= plt.subplots(3,7, figsize=(12,4))
titulos=list(dfGrilla["title"])
for ax, url, title in zip(axs.flatten(),imagenes,titulos):
    try:
        ax.imshow(io.imread(url,as_gray=True), cmap="gray")
        ax.set_title(title[:10])
        ax.axis("off")
    except:
        ax.imshow(io.imread("http://www.dagdrivarn.se/BILDER/stambilder/MISSING.jpg"))
        ax.axis("off")
```

```python
df["tag_count"]=df["tags"].apply(lambda x:x.count("|")+1)
df.head()
```

```python
df["Thumbs_up"]=df.apply(lambda x:True if x.likes>x.dislikes else False, axis=1)
df.head(5)
```

```python
with open("US_category_id.json", "r") as f:data=json.loads(f.read())
data
```

```python
jsonData=pd.json_normalize(data,record_path=["items"])
jsonData.head(5)
```

```python
jsonData=jsonData[["id","snippet.title"]]
jsonData=jsonData.rename(columns={"id":"category_id","snippet.title":"category_title"})
jsonData.head(5)
```

```python
jsonData["category_id"]=jsonData["category_id"].apply(pd.to_numeric)
```

```python
df=df.merge(jsonData, how="inner", left_on="category_id", right_on="category_id")
df=df[['title','channel_title','category_title','views','likes','dislikes','comment_count','description','tag_count','Thumbs_up']]
df.head(5)
```

```python
dfTop20=df.sort_values(by=["views"], ascending=False)[:20]
for i,m in enumerate(dfTop20.index):
    print(f"{i+1}: views:{dfTop20.views[m]}, title:{dfTop20.title[m]}")
    print("-"*80)
```

```python
fig, ax=plt.subplots()
ax.hist(df["tag_count"], color="green")
ax.set_title("Density of Tag Counts")
ax.set_ylabel("Frequency")
plt.show
```

```python
df["Thumbs_up"].value_counts().plot.bar().set_title("Thumbs Up Vs Down");
```

```python
fig, ax= plt.subplots(figsize=(15,4))
ax.bar(df["category_title"].unique().tolist(),height=df["category_title"].value_counts(ascending=False), color="red")
ax.tick_params(axis="x", labelrotation=90)
plt.show()
```

```python
fig, axs=plt.subplots(1,2,figsize=(15,8))
axs[0].scatter(df["likes"],df["dislikes"], color="magenta", marker="+", alpha=0.4, s=40)
axs[0].set_title("Likes vs Dislikes")
axs[0].set_xscale("log")
axs[0].set_yscale("log")

axs[1].scatter(df["views"],df["comment_count"], color="cyan", marker="*", alpha=0.7, s=70)
axs[1].set_title("Views vs Comments")
axs[1].set_xscale("log")
axs[1].set_yscale("log")

plt.show()
```

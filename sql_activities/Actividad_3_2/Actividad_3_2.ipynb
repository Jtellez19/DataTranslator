# 5.6- Obtener el 5º salario máximo de la tabla de empleados.


```python
df1 = pd.read_sql_query("""
                        with cte as
                        (
                            select
                            ROW_NUMBER() OVER(ORDER BY salary desc) AS orden,
                            salary
                            from employees
                        )
                        select salary from cte where orden = 5
                        
                        """,conn).head(5); df1
```
